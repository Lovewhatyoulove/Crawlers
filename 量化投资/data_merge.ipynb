```python
import pandas as pd

data1 = pd.read_csv('利润表.csv', encoding='gb2312', skiprows=[1, 2])
data2 = pd.read_csv('员工数.csv', encoding='gb2312', skiprows=[1, 2])
data3 = pd.read_csv('增发股票.csv', encoding='gb2312', skiprows=[1, 2])
data4 = pd.read_csv('现金流量表.csv', encoding='gb2312', skiprows=[1, 2])
data5 = pd.read_csv('资产负债表.csv', encoding='gb2312', skiprows=[1, 2])
data1['Stkcd'] = data1['Stkcd'].map(lambda x: '{:0>6d}'.format(x))
data2['Stkcd'] = data2['Stkcd'].map(lambda x: '{:0>6d}'.format(x))
data3['Stkcd'] = data3['Stkcd'].map(lambda x: '{:0>6d}'.format(x))
data4['Stkcd'] = data4['Stkcd'].map(lambda x: '{:0>6d}'.format(x))
data5['Stkcd'] = data5['Stkcd'].map(lambda x: '{:0>6d}'.format(x))
```

```python
data1.shape[0], data4.shape[0], data5.shape[0], data1['Stkcd'].unique().shape[0]
```

```python
data3['Accper'] = data3['Ailtdt'].map(lambda x: x[:4]+'-12-31')
```

```python
data = pd.merge(data1, data4, on=['Stkcd', 'Accper', 'Typrep'])
data = pd.merge(data, data5, on=['Stkcd', 'Accper', 'Typrep'])
data = pd.merge(data, data2, left_on=['Stkcd', 'Accper'], right_on=['Stkcd', 'Reptdt'], how='left')
data = pd.merge(data, data3, on=['Stkcd', 'Accper'], how='left')
data.drop('Typrep', inplace=True, axis=1)
data.head()
```

```python
data.drop(['Aistdt', 'Aiquty', 'Aitype', 'Reptdt'], inplace=True, axis=1)
```

```python
data['Ailtdt'].map(lambda x: x[:4] if pd.notnull(x) else 0)
```

```python
data['Newiss'] = data['Accper'].map(lambda x: x[:4]) == data['Ailtdt'].map(lambda x: x[:4] if pd.notnull(x) else 0)
```

```python
data['Newiss'] = data['Newiss'].astype(int)
data.drop(['B001100000', 'B001200000', 'Ailtdt'], axis=1, inplace=True)
data.drop_duplicates(inplace=True)
data.head()
```

```python
data.to_csv('财务数据.csv', index=False)
```
